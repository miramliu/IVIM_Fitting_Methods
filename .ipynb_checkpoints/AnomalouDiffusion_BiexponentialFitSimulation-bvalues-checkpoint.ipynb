```python
import numpy as np
import matplotlib
import matplotlib.pyplot as pl
%matplotlib inline
import csv
import time
import datetime

import scipy.optimize as op
import scipy.stats
from scipy.optimize import curve_fit
import scipy.io
from scipy.stats import rice

import random
import pickle
import seaborn as sns

import numpy as np
from scipy.stats import ttest_ind, ttest_ind_from_stats, wilcoxon,ttest_rel, pearsonr,shapiro,f_oneway, ranksums
from scipy.special import stdtr
import csv
import pandas as pd

from scipy.integrate import quad
import sys 
import os
sys.path.append(os.path.abspath("/Users/miraliu/Desktop/PostDocCode/IVIM_FItting_Methods_simulations"))

#from ML_UsefulFunctions import *
#from KidneyROIAnalyses import *

from scipy import special

from scipy.integrate import tplquad

pd.options.display.float_format = '{:.3f}'.format

from sklearn.metrics import r2_score
from sklearn.feature_selection import SequentialFeatureSelector

import warnings
warnings.simplefilter("ignore")


from EllipsoidalSimulationCode import *

from statistics import mean


def func(x,m,b):
    return m*x + b

def triexp_func(b, frac_fast,frac_med,frac_slow,diff_fast,diff_med,diff_slow):
    Data = frac_slow*np.exp(-b*diff_slow)+frac_med*np.exp(-b*diff_med)+frac_fast*np.exp(-b*diff_fast)
    normal = frac_slow+frac_med+frac_fast
    return Data/normal


## note here that diff_slow should be diff**gamma. instead it isn't. so it's slightly different. 

def anomalous_triexp_func(b, frac_fast,frac_med,frac_slow,diff_fast,diff_med,diff_slow, gamma_fast, gamma_med, gamma_slow):
    Data = frac_slow*np.exp(-(b*diff_slow)**gamma_slow)+frac_med*np.exp(-(b*diff_med)**gamma_med)+frac_fast*np.exp(-(b*diff_fast)**gamma_fast)
    normal = frac_slow+frac_med+frac_fast
    return Data/normal

def CalculatePE_fraction(df, trait):
    df[trait + ' PE'] = (df['Fit ' + trait]-df['true ' + trait])#*100/(df['true ' + trait])
    return df

def CalculatePE_diffusion(df, trait): #scale by 1000
    df[trait + ' PE'] = ((df['Fit ' +trait]/1000)-df['true ' + trait])#*100/(df['true ' + trait])
    return df

def blandalt_calc(FitfDs,TruefDs, name):
    mean_diff = np.mean(FitfDs-TruefDs)
    mean_diff_percent = 100*mean_diff/np.mean((FitfDs+TruefDs)/2)
    std_diff = np.std(FitfDs-TruefDs)
    std_diff_percent = 100*std_diff/np.mean((FitfDs+TruefDs)/2)
    lower_bound=mean_diff-(std_diff*1.96)
    upper_bound=mean_diff+(std_diff*1.96)
    lower_bound_percent=mean_diff_percent-(1.96*std_diff_percent)
    upper_bound_percent=mean_diff_percent+(1.96*std_diff_percent)
    print(f'{name} BA, {mean_diff:.3f}, [{lower_bound:.3f}-{upper_bound:.3f}]')
    #print(f'{name} %BA, {mean_diff_percent:.3f}%, [{lower_bound_percent:.3f}%-{upper_bound_percent:.3f}%]\n')
    return
    
def func(x,m,b):
    return m*x + b


def biexp_func(b, frac_fast,frac_slow,diff_fast,diff_slow):
    Data = frac_slow*np.exp(-b*diff_slow)+frac_fast*np.exp(-b*diff_fast)
    normal = frac_slow+frac_fast
    return Data/normal

def monoexp_func(b, frac, diff):
    Data = frac*np.exp(-b*diff)
    return Data

#Rician noise check

#SNR = 1, sigma = 1
#SNR = 10, sigma = .1
#SNR = 20, sigma = .045
#SNR = 50, sigma = .02
#SNR = 75, sigma = .013
#SNR = 100, sigma = .01
#SNR = 125, sigma = .008
#SNR = 150, sigma = .0065
def NoiseRice(I,SNR): # noise with rician distribution
    N = [] 
    sigma = 1/SNR
    #v = .79 # calculated from images on 6/11/19
    #sigma = .013 #(std of noise measured!)
    #v = 0.00434 #scaled = .79/182
    v = .005816 #scaled 6/27/19
    b = v/sigma
    r = rice.rvs(b, scale = sigma, size=len(I))
    for i in range(0,len(I)):
        N.append(I[i] +r[i]) #SNR = 1,10,22,100,150,inf
    return N


def ExportToDF(df, IdentifyingInfo, data):
    if isinstance(data, list):
        Export = data
        Export.insert(0, IdentifyingInfo)
        df.loc[ len(df),: ] = Export
    else:
        Export = data.tolist()
        Export.insert(0, IdentifyingInfo)
        df.loc[ len(df),: ] = Export
    return df

def GeneratefD(dataframe):
    if 'True fDstar' not in dataframe.columns:
        try: # if it is a fit-algorithm data frame, don't need to multiply by 1000
            dataframe['fDstar'] = dataframe['f']*dataframe['Dstar']

        except:
            dataframe['fDstar'] = 1000*dataframe['true fast fraction']*dataframe['true fast Diffusion']
    if 'True (1-f)D' not in dataframe.columns:
        try: # if it is a fit-algorithm data frame, don't need to multiply by 1000
            dataframe['(1-f)D'] = (1-dataframe['f'])*dataframe['D']

        except:
            dataframe['(1-f)D'] = 1000*dataframe['true slow fraction']*dataframe['true slow Diffusion']
    return dataframe
    
```

```python
## assign the variables to the dataframe columns of interest

def AnalyzeOutput(df_InputParameters, df_OutputParameters, AlgName, parameter):
    if parameter == 'fD':
        TruefastfDs=df_InputParameters['fDstar']
        FitfastfDs = df_OutputParameters['fDstar']
        TrueslowfDs=df_InputParameters['(1-f)D']
        FitslowfDs = df_OutputParameters['(1-f)D']
        xymin = -0.3
        xylim = 25
    elif parameter == 'f':
        TruefastfDs=df_InputParameters['true fast fraction']
        FitfastfDs = df_OutputParameters['f']
        TrueslowfDs=df_InputParameters['true slow fraction']
        FitslowfDs = 1-df_OutputParameters['f']
        xymin = 0
        xylim = 1
    elif parameter == 'D':
        TruefastfDs=df_InputParameters['true fast Diffusion']
        FitfastfDs = df_OutputParameters['Dstar']/1000
        TrueslowfDs=df_InputParameters['true slow Diffusion']
        FitslowfDs = df_OutputParameters['D']/1000
        xymin = 0 
        xylim = 0.1
    else:
        error('please input fD, f, or D')
        
    AllTruefD = list(TruefastfDs) + list(TrueslowfDs)
    allFitfD = list(FitfastfDs) + list(FitslowfDs)



    if parameter == 'f' or parameter == 'fD':
        pl.scatter(TruefastfDs, FitfastfDs, s=1)
        x = np.linspace(0,25,100)
        slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(TruefastfDs,FitfastfDs)
        r_value2 = r_value**2
        pl.plot(x,func(x,slope,intercept),label = f'y= {slope:.2f}x +{intercept: .1f}, $R^2$={r_value2:.2f}',color = 'grey') 
        pl.legend()
        pl.title(AlgName + ' fit: fast ' + parameter)
        pl.xlabel('Input fast ' + parameter)
        pl.ylabel('Output Fit fast ' + parameter)
        if parameter =='f':
            pl.xlim(xymin,0.5)
            pl.ylim(xymin,0.5)
        else:
            pl.xlim(xymin,xylim)
            pl.ylim(xymin,xylim)
        pl.show()

    else:
        pl.scatter(TruefastfDs, FitfastfDs,label= 'fast ' + parameter)
        pl.scatter(TrueslowfDs, FitslowfDs,label= 'slow ' + parameter)
    
        x = np.linspace(0,20,100)
        slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(AllTruefD,allFitfD)
        r_value2 = r_value**2
        pl.plot(x,func(x,slope,intercept),label = f'y= {slope:.2f}x +{intercept: .1f}, $R^2$={r_value2:.2f}',color = 'grey') 
        pl.legend()
        #pl.xlim(xymin,xylim)
        #pl.ylim(xymin,xylim)
        pl.title(AlgName + ' fit: ' + parameter)
        pl.show()


    pl.scatter(AllTruefD, allFitfD, s=1)
    x = np.linspace(0,25,100)
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(AllTruefD,allFitfD)
    r_value2 = r_value**2
    pl.plot(x,func(x,slope,intercept),label = f'y= {slope:.2f}x +{intercept: .1f}, $R^2$={r_value2:.2f}',color = 'grey') 
    pl.legend()
    pl.title(AlgName + ' fit: ' + parameter)
    pl.xlabel('Input ' + parameter)
    pl.ylabel('Output Fit ' + parameter)
    pl.xlim(xymin,xylim)
    pl.ylim(xymin,xylim)
    pl.show()

    
    

    print('relative percent difference')
    print(f'fast {parameter:s}, {200*np.median((FitfastfDs-TruefastfDs)/(TruefastfDs +FitfastfDs )):.5f}%')
    print(f'slow {parameter:s}, {200*np.median((FitslowfDs-TrueslowfDs)/(TrueslowfDs + FitslowfDs)):.5f}%')

    print('\nrelative abs percent difference')
    print(f'fast {parameter:s},  {200*np.median(abs(FitfastfDs-TruefastfDs)/(TruefastfDs +FitfastfDs )):.5f}%')
    print(f'slow {parameter:s}, {200*np.median(abs(FitslowfDs-TrueslowfDs)/(TrueslowfDs + FitslowfDs)):.5f}%')


    print('\nBland altman')
    blandalt_calc(FitfastfDs,TruefastfDs,'fast ' + parameter)
    blandalt_calc(FitslowfDs,TrueslowfDs,'slow ' + parameter)
    blandalt_calc(np.array(allFitfD),np.array(AllTruefD),'all '+ parameter)

```
